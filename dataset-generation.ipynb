## DeepL API

```python
from dotenv import dotenv_values

env = dotenv_values(".env")
```

```python
import deepl

translator = deepl.Translator(env.get("DEEPL"))
```

## Dataset

```python
#!wget "https://raw.githubusercontent.com/o19s/relevant-search-book/master/ipython/tmdb.json"
```

```python
import requests
import json

result = requests.get("https://raw.githubusercontent.com/o19s/relevant-search-book/master/ipython/tmdb.json")
ds = json.loads(result.text)
```

```python
import pandas as pd

df = pd.DataFrame(ds)
df.head()
```

```python
df = df.T
df.head()
```

```python
df.columns
```

```python
df = df[["title", "overview"]].dropna()
df.head()
```

```python
# translate
df["ger"] = df["overview"].apply(lambda text: translator.translate_text(text, target_lang="DE"))
df.head()
```

```python
df.to_csv("dataset.csv")
```
